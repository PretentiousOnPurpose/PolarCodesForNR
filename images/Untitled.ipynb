```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
x = mpimg.imread("test_image.png")
```

```python
x1 = x.reshape(-1,) * 255.0
```

```python
with open("test_im_bin", "wb") as fp:    
    for i in range(len(x1)):
        fp.write(int(x1[i]).to_bytes(1, byteorder='big'))
```

```python
y = np.zeros((len(x1), 1))

with open("test_im_rec_bin", "rb") as fp:
    for i in range(len(x1)):
        y[i] = int.from_bytes(fp.read(1), byteorder='big') / 255.0
```
